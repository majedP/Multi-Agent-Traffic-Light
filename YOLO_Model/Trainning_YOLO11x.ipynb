# **Step 4: Writing The YAML File to train the YOLO Model**

```python

yaml_content = """
path: /content/vehicle dataset
train: /content/vehicle dataset/train
val: /content/vehicle dataset/valid
nc: 6  # Number of classes
names: ['car', 'threewheel', 'bus', 'truck', 'motorbike', 'van']  # List of class names in order
"""


with open('/content/vehicle dataset/cars.yaml', 'w') as file:
    file.write(yaml_content)

print("YAML file created successfully!")
```

# **Step 5: Importing YOLO11x Model to Train.**

```python
model = YOLO("yolo11x.pt")
```

# **Step 6: Printing the model layers to freeze it.**

```python
print(model)
```

# **Step 7: Trainning YOLO11x by freezing 22 Layers.**

```python


# Start training the model
model.train(data='/content/vehicle dataset/cars.yaml', epochs=5, imgsz=640, batch=16, freeze=22)
```

# **Step 8: Saving the model.**

```python
model.save('/content/trained11modelx.pt')
```
